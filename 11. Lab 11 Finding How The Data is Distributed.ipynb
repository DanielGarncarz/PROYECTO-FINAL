### Step 1: Import Libraries and Load Data

- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
print(df.info())
print(df.dtypes)
df.describe()
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
df.isnull().sum().sort_values(ascending = False)
```

```python
for columna in df.columns:
    if df[columna].isnull().any():
        if df[columna].dtype == 'object':
            moda = df[columna].mode()[0]
            df[columna] = df[columna].fillna(moda)
        elif pd.api.types.is_numeric_dtype(df[columna]):
            media = df[columna].mean()
            df[columna] = df[columna].fillna(media)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
print("Distribución de Employment:")
print(df['Employment'].value_counts(dropna=False).head())

print("Distribución de Job Satisfaction:")
print(df['JobSat'].value_counts(dropna=False).head())

print("Distribución de Años de Experiencia Profesional:")
print(df['YearsCodePro'].value_counts(dropna=False).head())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.

```python
X = df['JobSat'].value_counts(dropna=False)
```

```python
plt.figure(figsize=(8,6))
plt.pie(X, labels= X.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Distribution of Job Satisfaction")
plt.axis('equal')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
HaveWorked = df["LanguageHaveWorkedWith"].str.split(";").explode()
WantToWork = df["LanguageWantToWorkWith"].str.split(";").explode()
HaveWorked_counts = HaveWorked.value_counts()
WantToWork_counts = WantToWork.value_counts()

Comparative = pd.concat([HaveWorked_counts, WantToWork_counts], axis=1, keys=['HaveWorkedWith', 'WantToWorkWith']).fillna(0)


Comparative.head(10).plot(kind='bar', figsize=(14,6))
plt.title('Comparación de lenguajes: Trabajado vs Querer Trabajar')
plt.ylabel('Cantidad de respuestas')
plt.xticks(rotation=90)
plt.legend(["Trabajado con", "Quiere trabajar con"])
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
new_df = df[["RemoteWork","Country"]]
country_counts = new_df["Country"].value_counts()
top_countries = country_counts[country_counts > 600].index
new_df_top = new_df[new_df["Country"].isin(top_countries)]
data = pd.crosstab(new_df_top["Country"],new_df_top["RemoteWork"])

plt.figure(figsize=(16,10))
sns.heatmap(data,annot=True, fmt='d', cmap="YlGnBu")
plt.title("Distribución de trabajo remoto por país")
plt.xlabel("Trabajo remoto")
plt.ylabel("País")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install spicy
from scipy.stats import pearsonr
```

```python
df["YearsCodePro"].replace('Less than 1 year', 0, inplace=True)
df["YearsCodePro"].replace('More than 50 years', 50, inplace=True)

df["YearsCodePro"] = df["YearsCodePro"].astype(float)

new_df = df[["JobSat", "YearsCodePro"]].copy()

new_df["JobSat"] = pd.to_numeric(new_df["JobSat"], errors='coerce')

new_df = new_df.dropna()

pearson_corr, _ = pearsonr(new_df["JobSat"], new_df["YearsCodePro"])
print("Correlación de Pearson:", round(pearson_corr, 3))

plt.figure(figsize=(8,6))
sns.regplot(x='YearsCodePro', y='JobSat', data=new_df, line_kws={'color':'red'})
plt.title('Relación entre Años de Experiencia y Satisfacción Laboral')
plt.xlabel('Años de Experiencia Profesional (YearsCodePro)')
plt.ylabel('Satisfacción Laboral (JobSat)')
plt.grid(True)
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
crosstab = pd.crosstab(df['EdLevel'].head(10), df['Employment'].head(10))

crosstab.head()

crosstab.plot(kind='bar', stacked=True, figsize=(12,8), colormap='tab20')

plt.title('Relación entre Nivel de Educación y Estado Laboral')
plt.xlabel('Nivel de Educación')
plt.ylabel('Cantidad de Personas')
plt.legend(title='Estado Laboral', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
