### Step 2: Load the Dataset


```python
df = pd.read_csv("Lab 12.csv")

df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
X = df["ConvertedCompYearly"].dropna()

Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

Y = X[(X >= limite_inferior) & (X <= limite_superior)]

plt.figure(figsize=(12,6))
sns.histplot(Y,kde=True, bins=10, color='skyblue')
plt.title("DistribuciÃ³n del salario anual")
plt.xlabel("Salario anual en USD",fontsize=12)
plt.ylabel("Cantidad de personas",fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
full_time = df[df["Employment"] == "Employed, full-time"]
full_time = full_time["ConvertedCompYearly"].dropna()
mediana_salario = full_time.median()

print("La mediana del salario anual para empleados a tiempo completo es: $",mediana_salario)
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
df_filtered = df[["Country", "ConvertedCompYearly"]].dropna()

upper_limit = df_filtered["ConvertedCompYearly"].quantile(0.95)
df_filtered = df_filtered[df_filtered["ConvertedCompYearly"] <= upper_limit]

top_countries = df_filtered["Country"].value_counts().head(10).index
df_top = df_filtered[df_filtered["Country"].isin(top_countries)]

plt.figure(figsize=(14, 8))
sns.boxplot(x="ConvertedCompYearly", y="Country", data=df_top, palette="Set3")
plt.title("DistribuciÃ³n del salario anual por paÃ­s (sin outliers extremos)")
plt.xlabel("Salario anual en USD")
plt.ylabel("PaÃ­s")
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
new_df = df[df["ConvertedCompYearly"].notna()]

Q1 = new_df["ConvertedCompYearly"].quantile(0.25)
Q3 = new_df["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

new_df = new_df[
    (new_df["ConvertedCompYearly"] >= limite_inferior) & 
    (new_df["ConvertedCompYearly"] <= limite_superior)
]

new_df.head()
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
corr_df = new_df[["ConvertedCompYearly","WorkExp","JobSatPoints_1"]].copy()

corr_df["ConvertedCompYearly"] = corr_df["ConvertedCompYearly"].fillna(corr_df["ConvertedCompYearly"].mean())
corr_df["WorkExp"] = corr_df["WorkExp"].fillna(corr_df["WorkExp"].mean())
corr_df["JobSatPoints_1"] = corr_df["JobSatPoints_1"].fillna(corr_df["JobSatPoints_1"].mean())

matriz = corr_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(matriz, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlaciones entre salario, experiencia y satisfacciÃ³n laboral", fontsize=14)
plt.tight_layout()
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
sns.scatterplot(x="WorkExp", y="ConvertedCompYearly", data=corr_df, alpha=0.5)
plt.title("Salario anual vs. AÃ±os de experiencia")
plt.xlabel("AÃ±os de experiencia")
plt.ylabel("Salario anual (USD)")

plt.subplot(1, 2, 2)
sns.scatterplot(x="JobSatPoints_1", y="ConvertedCompYearly", data=corr_df, alpha=0.5)
plt.title("Salario anual vs. SatisfacciÃ³n laboral")
plt.xlabel("Nivel de satisfacciÃ³n laboral")
plt.ylabel("Salario anual (USD)")

plt.tight_layout()
plt.show()
```
