#### Step 1: Load the dataset


```python
df = pd.read_csv("Lab 13.csv")
df.head()
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
new_df = df[["AgeNumeric", "JobSatPoints_6"]].dropna()

plt.figure(figsize=(10, 6))
sns.regplot(data=new_df, x="AgeNumeric", y="JobSatPoints_6", scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("GrÃ¡fico de dispersiÃ³n de Edad y SatisfacciÃ³n laboral")
plt.xlabel("Edad")
plt.ylabel("SatisfacciÃ³n laboral")
plt.grid(True)
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
new_df = df[["ConvertedCompYearly", "JobSatPoints_6"]].dropna()

Q1 = df["ConvertedCompYearly"].quantile(0.25)
Q3 = df["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

new_df = df[(df["ConvertedCompYearly"] >= limite_inferior) & (df["ConvertedCompYearly"] <= limite_superior)]

plt.figure(figsize=(10, 6))
sns.regplot(data=new_df, x="ConvertedCompYearly", y="JobSatPoints_6", scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("GrÃ¡fico de dispersiÃ³n de CompensaciÃ³n anual y SatisfacciÃ³n laboral")
plt.xlabel("CompensaciÃ³n anual")
plt.ylabel("SatisfacciÃ³n laboral")
plt.grid(True)
plt.show()
```

### Task 2: Enhancing Scatter Plots

Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
bubble_df = df[["ConvertedCompYearly", "JobSatPoints_6", "AgeNumeric"]].dropna()

Q1 = bubble_df["ConvertedCompYearly"].quantile(0.25)
Q3 = bubble_df["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

new_df = bubble_df[(bubble_df["ConvertedCompYearly"] >= limite_inferior) & (bubble_df["ConvertedCompYearly"] <= limite_superior)]

tamaÃ±o_burbujas = new_df["AgeNumeric"] * 10

plt.figure(figsize=(12, 7))
plt.scatter(new_df["ConvertedCompYearly"], new_df["JobSatPoints_6"], s=tamaÃ±o_burbujas, alpha=0.5, c=new_df["AgeNumeric"], cmap='viridis', edgecolors='w', linewidths=0.5)

plt.title("GrÃ¡fico de burbujas: CompensaciÃ³n anual vs SatisfacciÃ³n laboral por Edad")
plt.xlabel("CompensaciÃ³n anual")
plt.ylabel("SatisfacciÃ³n laboral")
plt.grid(True)
plt.colorbar(label='Edad')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
new_df = df[["LanguageHaveWorkedWith", "JobSatPoints_6"]].dropna()
new_df["LanguageHaveWorkedWith"] = new_df["LanguageHaveWorkedWith"].str.split(";")
X = new_df.explode("LanguageHaveWorkedWith")
X.rename(columns={"LanguageHaveWorkedWith": "Language"}, inplace=True)

Y = X.groupby("Language")["JobSatPoints_6"].mean().reset_index()
Z = X["Language"].value_counts().reset_index()
Z.columns = ["Language", "Count"]

lang_plot = pd.merge(Y, Z, on="Language")

plt.figure(figsize=(12, 7))
plt.scatter(lang_plot["Language"], lang_plot["JobSatPoints_6"], s=lang_plot["Count"], alpha=0.6, c='teal', edgecolors='w')

plt.title("SatisfacciÃ³n laboral segÃºn lenguaje de programaciÃ³n")
plt.xlabel("Lenguaje de programaciÃ³n")
plt.ylabel("SatisfacciÃ³n laboral promedio")
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
new_df = df[["ConvertedCompYearly", "JobSatPoints_6", "TipoEmpleoAgrupado"]].dropna()

Q1 = new_df["ConvertedCompYearly"].quantile(0.25)
Q3 = new_df["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

new_df = new_df[(new_df["ConvertedCompYearly"] >= limite_inferior) & (new_df["ConvertedCompYearly"] <= limite_superior)]

plt.figure(figsize=(12, 7))
sns.scatterplot(data=new_df, x="ConvertedCompYearly", y="JobSatPoints_6", hue="TipoEmpleoAgrupado", alpha=0.6, palette="tab10")

plt.title("RelaciÃ³n entre CompensaciÃ³n anual y SatisfacciÃ³n laboral por tipo de empleo")
plt.xlabel("CompensaciÃ³n anual")
plt.ylabel("SatisfacciÃ³n laboral")
plt.grid(True)
plt.legend(title="Tipo de empleo", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
new_df = df[["YearsCodePro", "Age", "AgeNumeric", "Country"]].dropna()

top_paises = new_df["Country"].value_counts().head(5).index
new_df = new_df[new_df["Country"].isin(top_paises)]

plt.figure(figsize=(12, 7))
sns.scatterplot(data=new_df, x="AgeNumeric", y="YearsCodePro", hue="Country", style="Age", palette="Set2", alpha=0.6)

plt.title("Experiencia laboral por edad y paÃ­s")
plt.xlabel("Edad")
plt.ylabel("AÃ±os de experiencia profesional")
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
