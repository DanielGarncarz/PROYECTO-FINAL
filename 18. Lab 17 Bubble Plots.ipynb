### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
new_df = df[["AgeNumeric", "JobSat","SOPartFreq"]].dropna()
top_freq = new_df["SOPartFreq"].value_counts().head(5).index
new_df = new_df[new_df["SOPartFreq"].isin(top_freq)]

plt.figure(figsize=(12, 7))
plt.scatter(x=new_df["AgeNumeric"], y=new_df["SOPartFreq"], s=new_df["JobSat"] * 50, alpha=0.5, edgecolors="w", linewidths=0.5, c=new_df["JobSat"], cmap="viridis")
plt.title("Relación entre Edad y Frecuencia de participación")
plt.xlabel("Edad")
plt.ylabel("Frecuencia de participación")
plt.grid(True)
plt.colorbar(label="Nivel de Satisfacción laboral")
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
new_df = df[["AgeNumeric", "JobSat","ConvertedCompYearly"]].dropna()

Q1 = new_df["ConvertedCompYearly"].quantile(0.25)
Q3 = new_df["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

new_df = new_df[(new_df["ConvertedCompYearly"] >= limite_inferior) & (new_df["ConvertedCompYearly"] <= limite_superior)]

plt.figure(figsize=(12, 7))
plt.scatter(x=new_df["ConvertedCompYearly"], y=new_df["JobSat"], s=new_df["AgeNumeric"] * 20, alpha=0.5, edgecolors="w", linewidths=0.5, c=new_df["AgeNumeric"], cmap="viridis")
plt.title("Relación entre Compensación anual, Satisfacción laboral y Edad")
plt.ylabel("Satisfacción laboral")
plt.xlabel("Compensación anual")
plt.grid(True)
plt.colorbar(label="Edad")
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
new_df = df[["Age", "LanguageHaveWorkedWith"]].dropna()
top_freq = new_df["LanguageHaveWorkedWith"].value_counts(ascending = False).head(10).index
new_df = new_df[new_df["LanguageHaveWorkedWith"].isin(top_freq)]

df_expanded = new_df.assign(Language=new_df["LanguageHaveWorkedWith"].str.split(";")).explode("Language")
language_counts = df_expanded.groupby(["Age", "Language"]).size().reset_index(name="Frequency")

plt.figure(figsize=(14, 8))
plt.scatter(x=language_counts["Age"], y=language_counts["Language"], s=language_counts["Frequency"] * 2, alpha=0.6, edgecolors="w", linewidths=0.5,
            c=language_counts["Frequency"], cmap="Blues")

plt.title("Popularidad de lenguajes de programación por grupo etario")
plt.xlabel("Grupo de Edad")
plt.ylabel("Lenguaje de Programación")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.colorbar(label="Frecuencia de uso")
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
new_df = df[["JobSat", "DatabaseWantToWorkWith"]].dropna()
top_freq = new_df["DatabaseWantToWorkWith"].value_counts(ascending = False).head(10).index
new_df = new_df[new_df["DatabaseWantToWorkWith"].isin(top_freq)]

df_expanded = new_df.assign(Language=new_df["DatabaseWantToWorkWith"].str.split(";")).explode("Language")
language_counts = df_expanded.groupby(["JobSat", "Language"]).size().reset_index(name="Frequency")

plt.figure(figsize=(14, 8))
plt.scatter(x=language_counts["JobSat"], y=language_counts["Language"], s=language_counts["Frequency"] * 2, alpha=0.6, edgecolors="w", linewidths=0.5,
            c=language_counts["Frequency"], cmap="Blues")

plt.title("Relación entre bases de datos preferidas y satisfacción laboral")
plt.xlabel("Satisfacción laboral")
plt.ylabel("Base de datos")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.colorbar(label="Cantidad de encuestados")
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
new_df = df[["DevType", "JobSat","ConvertedCompYearly"]].dropna()

Q1 = new_df["ConvertedCompYearly"].quantile(0.25)
Q3 = new_df["ConvertedCompYearly"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

new_df = new_df[(new_df["ConvertedCompYearly"] >= limite_inferior) & (new_df["ConvertedCompYearly"] <= limite_superior)]

top_freq = new_df["DevType"].value_counts(ascending = False).head(10).index
new_df = new_df[new_df["DevType"].isin(top_freq)]

plt.figure(figsize=(12, 7))
plt.scatter(x=new_df["ConvertedCompYearly"], y=new_df["DevType"], s=new_df["JobSat"] * 20, alpha=0.5, edgecolors="w", linewidths=0.5, c=new_df["JobSat"], cmap="viridis")
plt.title("Relación entre Compensación anual, Rol de desarrollador y Satisfacción laboral")
plt.ylabel("Rol de desarrollador")
plt.xlabel("Compensación anual")
plt.grid(True)
plt.colorbar(label="Satisfacción laboral")
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
new_df = df[["Age", "NEWCollabToolsHaveWorkedWith"]].dropna()
top_freq = new_df["NEWCollabToolsHaveWorkedWith"].value_counts(ascending = False).head(10).index
new_df = new_df[new_df["NEWCollabToolsHaveWorkedWith"].isin(top_freq)]

df_expanded = new_df.assign(Tool=new_df["NEWCollabToolsHaveWorkedWith"].str.split(";")).explode("Tool")
Tool_counts = df_expanded.groupby(["Age", "Tool"]).size().reset_index(name="Frequency")

plt.figure(figsize=(14, 8))
plt.scatter(x=Tool_counts["Age"], y=Tool_counts["Tool"], s=Tool_counts["Frequency"] * 2, alpha=0.6, edgecolors="w", linewidths=0.5,
            c=Tool_counts["Frequency"], cmap="Blues")

plt.title("Herramientas de colaboración utilizadas por grupo etario")
plt.xlabel("Grupo de Edad")
plt.ylabel("Herramienta")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.colorbar(label="Frecuencia de uso")
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
df_web = df[["WebframeWantToWorkWith", "JobSat"]].dropna()
df_web = df_web.assign(Framework=df_web["WebframeWantToWorkWith"].str.split(";")).explode("Framework")
grouped = df_web.groupby(["Framework", "JobSat"]).size().reset_index(name="Count")

plt.figure(figsize=(14, 8))
plt.scatter(x=grouped["Framework"], y=grouped["JobSat"], s=grouped["Count"] * 2, alpha=0.6, c=grouped["Count"], cmap="viridis", edgecolors="w", linewidths=0.5)

plt.title("Relación entre frameworks web preferidos y satisfacción laboral")
plt.xlabel("Framework Web Preferido")
plt.ylabel("Nivel de Satisfacción Laboral")
plt.xticks(rotation=45, ha="right")
plt.grid(True, linestyle="--", alpha=0.5)
plt.colorbar(label="Cantidad de personas")
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
new_df = df[["LanguageAdmired", "Country"]].dropna()
new_df = new_df.assign(Language=new_df["LanguageAdmired"].str.split(";")).explode("Language")
grouped = new_df.groupby(["Language", "Country"]).size().reset_index(name="Count")
top_countries = grouped.groupby("Country")["Count"].sum().nlargest(10).index
top_languages = grouped.groupby("Language")["Count"].sum().nlargest(10).index
filtered = grouped[grouped["Country"].isin(top_countries) & grouped["Language"].isin(top_languages)]

plt.figure(figsize=(14, 8))
plt.scatter(x=filtered["Language"], y=filtered["Country"], s=filtered["Count"] * 2, alpha=0.6, c=filtered["Count"], cmap="viridis", edgecolors="w", linewidths=0.5)
plt.title("Top 10 tecnologías admiradas por país (Top 10 países)")
plt.xlabel("Tecnologías admiradas")
plt.ylabel("País")
plt.xticks(rotation=45, ha="right")
plt.grid(True, linestyle="--", alpha=0.5)
plt.colorbar(label="Cantidad de personas")
plt.tight_layout()
plt.show()
```
